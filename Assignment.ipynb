```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
from sklearn.preprocessing import normalize
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
df = df.iloc[:, 1:]
df.head()
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

X = df.drop('Class', axis=1)
y = df['Class']


oversampler = RandomOverSampler()



x_over, y_over = oversampler.fit_resample(X, y)


print(y_over.value_counts())
```

```python
r_df = pd.concat([x_over, y_over], axis=1)
r_df
```

```python
n = int((1.96*1.96 * 0.5*0.5)/(0.05**2))
sampled_df = r_df.sample(n=n, random_state=42)
print(sampled_df)
```

```python
# systematic sampling

interval = 3
systematic_df = r_df.iloc[::interval]
systematic_df
```

```python


n = int((1.96*1.96 * 0.5*0.5)/((0.05)**2))

strata = r_df.groupby('Class')


stratified_df = strata.apply(lambda x: x.sample(n))

stratified_df
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8, random_state=40).fit(r_df)
cluster_a = kmeans.labels_
selected_clusters = [0, 2, 4]
cluster_series = pd.Series(cluster_a)
cluster_df = r_df[cluster_series.isin(selected_clusters)]
```

```python
# convenience sampling
convenience_df = pd.concat([r_df.head(350), r_df.tail(350)])
convenience_df
```

```python
from sklearn.tree import DecisionTreeClassifier as DT
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.linear_model import LogisticRegression as LR
from sklearn.naive_bayes import GaussianNB as NB
from sklearn.tree import ExtraTreeClassifier as ET
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as ac
import numpy as np
tr_simple,t_simple,ytr_simple,yt_simple=train_test_split(sampled_df.drop(["Class"],axis=1),sampled_df.Class,test_size=0.2)
tr_systematic, t_systematic, ytr_systematic, yt_systematic = train_test_split(systematic_df.drop(["Class"], axis=1), systematic_df.Class, test_size=0.2)

tr_stratified, t_stratified, ytr_stratified, yt_stratified = train_test_split(stratified_df.drop(["Class"], axis=1), stratified_df.Class, test_size=0.2)
tr_cluster, t_cluster, ytr_cluster, yt_cluster = train_test_split(cluster_df.drop(["Class"], axis=1), cluster_df.Class, test_size=0.2)
tr_convenience, t_convenience, ytr_convenience, yt_convenience = train_test_split(convenience_df.drop(["Class"], axis=1), convenience_df.Class, test_size=0.2)
```

```python
# Simple
simple = pd.DataFrame(columns=['DT', 'RF', 'LR', 'ET', 'NB'])
li = []
dt = DT()
dt.fit(tr_simple, ytr_simple)
p = dt.predict(t_simple)
li.append(ac(yt_simple, p))

rf = RF()
rf.fit(tr_simple, ytr_simple)
p = rf.predict(t_simple)
li.append(ac(yt_simple, p))


lr = LR()
lr.fit(tr_simple, ytr_simple)
p = lr.predict(t_simple)
li.append(ac(yt_simple, p))


et = ET()
et.fit(tr_simple, ytr_simple)
p = et.predict(t_simple)
li.append(ac(yt_simple, p))

nb = NB()
nb.fit(tr_simple, ytr_simple)
p = nb.predict(t_simple)
li.append(ac(yt_simple, p))

simple.loc[len(simple.index)] = li
```
